<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
1. Perbedaan antara list dan series antara lain:
  - series ditampilkan (print) beserta dengan indexnya, sedangkan list hanya menampilkan value saja
  - series tidak harus mempunyai index bertipe integer atau bisa berupa sting seperti dictionary, sedangkan list hanya bisa memiliki index integer
  - series bisa melakukan slicing pada data di dalamnya seperti list, sedangkan dictionary tidak bisa langsung di-slice
  - series bisa memiliki key (index) yang duplikat, sedangkan dictionary tidak

2. Perbedaan antara dataframe dengan series adalah dataframe merupakan objek 2 dimensi dan memiliki 2 index, yaitu index baris dan index kolom. Sedangkan list hanya objek 1 dimensi dan memiliki 1 index saja. Berdasarkan perbedaan tersebut, dataframe dapat menyimpan data dengan lebih terstruktur layaknya tabel dibandingkan dengan series.

```python
# code here
series_belanja = pd.Series(["Bawang Merah","Bawang Putih","Cabe Merah","Cabe Rawit","Beras","Minyak Goreng"])
print(series_belanja)
print("\n",type(series_belanja))
```

<h2>Soal 4: Pemeriksaan Data Sederhana</h2>

Di soal no 2 kamu telah membaca data tentang diamonds. pada soal ini cobalah suatu metode dari pandas untuk mengetahui beberapa data pertama dan beberapa data terakhir. kemudian sebutkan nilai dari:
- baris pertama dari column price
- baris terakhir dari column color

tunjukan cara kalian menemukan nilai tersebut di block di bawah ini.

```python
# code here
# Menampilkan  baris pertama dataframe dengan method .head() (nilai default 5 baris)
print(df.head())
# Menampilkan baris terakhir datafram dengan method .tail() (nilai default 5 baris)
print(df.tail())

# Slicing untuk kolom price dan menampilkan hanya baris paling pertama dengan method .head()
print("Baris paling pertama dari kolom price memiliki nilai", df['price'].head(1).to_string(index=False))
# Slicing untuk kolom color dan menampilkan hanya baris paling terakhir dengan method .tail()
print("Baris paling pertama dari kolom price memiliki nilai", df['color'].tail(1).to_string(index=False))
```

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
# Mengetahui deskriptif statistik dari suatu data dengan pandas
df.describe()
```

Jawab:
1. Nilai rata-rata dari kolom price adalah :  3932.799722	
2. Nilai standar deviasi dari kolom column depth adalah :  1.432621
3. Nilai maximum dari kolom carat adalah :  5.01

```python
# Mengetahui nilai rata-rata dari kolom price
print("rata-rata dari kolom price adalah : ", df['price'].describe()['mean'])
# Mengetahui standar deviasi dari kolom depth
print("standar deviasi dari kolom column depth adalah : ", df['depth'].describe()['std'])
# Mengetahui nilai maximum dari kolom carat
print("nilai maximum dari kolom carat adalah : ", df['carat'].describe()['max'])
```
