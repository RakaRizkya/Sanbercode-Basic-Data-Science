<h2>Soal 1: Comparison Operator</h2>

- Berikan contoh comparison dari tipe data string
- Berikan contoh comparison dari tipe data boolean
- Berikan contoh comparison dari tipe data integer

```python
# Comparison untuk tipe data string dengan dua variabel a dan b
a = "badak"
b = "gajah"
print (a == b)
 
# Comparison untuk tipe data boolean dengan variabel c dan d
c = True
d = False
print(c > d)
 
# Comparison untuk tipe data integer dengan variabel e dan f
e = 12
f = 67
print(e != f)
```

<h2>Soal 2: Boolean Comparison</h2>

- Berikan contoh gabungan multiple comparison menggunkan Boolean Comparisin 'and'
- Berikan contoh gabungan multiple comparison menggunkan Boolean Comparisin 'or'
- Berikan contoh gabungan multiple comparison menggunkan Boolean Comparisin 'not'

```python
# Deklarasi variabel-variabel yang akan digunakan
angka1 = 13.9
angka2 = 77.5
hewan1 = 'anjing'
hewan2 = 'kucing'
# Multiple comparison menggunkan Boolean Comparison 'and'
print(angka1 > angka2 and hewan1 != hewan2)               # Comparison pertama = False, Comparison kedua = True
# Multiple comparison menggunkan Boolean Comparison 'or'
print(angka1 > angka2 or hewan1 != hewan2)                # Comparison pertama = False, Comparison kedua = True
# Multiple comparison menggunkan Boolean Comparison 'not'
print(not(angka1 > angka2))                               # Nilai akhir True karena comparison bernilai False
```

<h2>Soal 4: Comparison Operator dengan fungsi</h2>

Buatlah sebuah fungsi yang menerima satu argument bertipe data numeric dan menghasilkan sebuah return sebagai berikut :
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan ganjil
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 2 sampai 5 (2 dan 5 termasuk)
- menghasilkan return 'Aneh' jika nilai dari argument tersebut adalah bilangan genap dan diantara nilai 6 sampai 20 (6 dan 20 termasuk)
- menghasilkan return 'Tidak Aneh' jika nilai dari argument tersebut adalah bilangan genap dan lebih besari dari 20

```python
angka1 = 3
angka2 = 2
angka3 = 20
angka4 = 22
# Buat fungsi
def komparasi(masukan):
  if masukan % 2 != 0:
    print('Aneh')
  elif masukan % 2 == 0 and masukan in range(2, 6):
    print('Tidak Aneh')
  elif masukan % 2 == 0 and masukan in range(6, 21):
    print('Aneh')
  elif masukan % 2 == 0 and masukan > 20:
    print('Tidak Aneh')

komparasi(angka1)
komparasi(angka2)
komparasi(angka3)
komparasi(angka4)
```

<h2>Soal 5: While Loop dan For Loop</h2>

- Apa perbedaan while loop dan for loop?
- Berikan contoh sederhana cara menggunakan while loop dan for loop

***JAWABAN : Perbedaan while loop dan for loop adalah for loop biasanya digunakan untuk mengulangi kode yang sudah diketahui banyak perulangannya, sedangkan while loop terus menerus mengulangi kode sampai memenuhi suatu kondisi yang membuat dia harus berhenti***

```python
# Tulis Kode
# Contoh penggunaan while loop untuk menghitung bilangan genap antara 2 sampai 20
bilangan = 2
print("Contoh perulangan dengan while :")
while bilangan <= 20:
  print(bilangan, end=' ')
  bilangan += 2
# Contoh penggunaan for loop untuk menghitung bilangan ganjil antara 1 sampai 19
print("\nContoh perulangan dengan for :")
for i in range(1, 20, 2):
  print(i, end=' ')
```

<h2>Soal 6: While Loop</h2>
    
Buatlah sebuah code while loop sebagai berikut :
- Buatlah suatu fungsi yang merima satu input nilai numeric
- Dalam fungsi, Buat sebuah while-loop dengan melakukan comparasi terhadap inputan tersebut, dimana looping/iterasi terus berjalan apabila nilai inputan tidak sama dengan 0
- Dalam setiap iterasi update nilai variable inputan itu, jika input lebih dari 0, maka kurangi variabel tersebut sebesar 1, selain itu tambahkan 1
- print nilai variable input dalam setiap awal iterasi

```python
a = 10

def fungsi_while(masukan):
    while masukan >= 1:
      print(masukan)
      masukan -= 1

fungsi_while(a)
```
