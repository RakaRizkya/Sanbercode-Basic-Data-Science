-**Axis** adalah suatu area di dalam figure dimana data akan di-plot

<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)
```

```python
# Membuat figure 1 dan axis 1
fig1 = plt.figure()
ax1 = fig1.add_subplot()

# Membuat figure 2 dan axis 2 secara explicit
fig2 = plt.figure()
ax2 = fig2.add_subplot()

# Menambah dan menampilkan plot ke dalam axis di figure 2
ax2.plot(x, y)
plt.show()
```

```python
# Membuat figure dan axis dengan data x dan y secara implicit
plt.plot(x, y)
plt.show()
```

<h2>Soal 3: Memasukan multiple data ke dalam satu axis</h2>

- Plot ketiga data berikut ke dalam satu axis, dan gunakan seaborn style
- save data tersebut dengan nama file bebas

```python
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
y2 = np.sin(x)
y3 = np.sin(2*x)
```

```python
# Menerapkan style seaborn pada figure dan axis
plt.style.use('seaborn')

# Membuat figure dan axis secara explicit
fig, ax = plt.subplots()

ax.plot(x, y, x, y2, x, y3)
plt.show()

# Menyimpan hasil plot dengan nama gambar.png
fig.savefig('gambar.png')
```
